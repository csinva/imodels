```python
import warnings
%load_ext autoreload
%autoreload 2
import numpy as np
import pandas as pd
from autogluon.tabular import TabularDataset, TabularPredictor
```

**load data**

```python
# train data
train_data = TabularDataset('https://autogluon.s3.amazonaws.com/datasets/Inc/train.csv')
subsample_size = 100  # subsample subset of data for faster demo, try setting this to much larger values
train_data = train_data.sample(n=subsample_size, random_state=0)
# train_data.head()

# test data
test_data = TabularDataset('https://autogluon.s3.amazonaws.com/datasets/Inc/test.csv')
label = 'class'
y_test = test_data[label]  # values to predict
test_data_nolab = test_data.drop(columns=[label])  # delete label column to prove we're not cheating
# test_data_nolab.head()
# print("Summary of class variable: \n", train_data[label].describe())
```

# Fit models using interpretable preset

**fit using interpretable preset**

```python
import warnings
warnings.simplefilter('ignore')
predictor = TabularPredictor(label=label).fit(train_data, presets='interpretable', time_limit=8, verbosity=2)
```

```python
predictor.leaderboard(silent=True)
```

# tune custom models

```python
from autogluon.core.space import Categorical, Int, Real
from autogluon.tabular.configs.hyperparameter_configs import get_hyperparameter_config

custom_hyperparameters_hpo = {
    RuleFitModel: {
        'tree_size': Int(lower=5, upper=30),
        'max_rules': Int(lower=2, upper=15),
    },
    GreedyTreeModel: {
        'max_leaf_nodes': Int(lower=2, upper=30),
        'ccp_alpha': Real(lower=0, upper=5),
    }
}
# Hyperparameter tune CustomRandomForestModel for 20 seconds
predictor = TabularPredictor(label=label).fit(train_data,
                                              hyperparameters=custom_hyperparameters_hpo,
                                              hyperparameter_tune_kwargs='auto',  # enables HPO
                                              time_limit=60)
```

```python
leaderboard = predictor.leaderboard(silent=True)
leaderboard
```

# test individual model

```python
# Separate features and labels
X = train_data.drop(columns=[label])
y = train_data[label]
X_test = test_data.drop(columns=[label])
y_test = test_data[label]

# basic model fitting
save_path = 'AutogluonModels'  # specifies folder to store trained models
predictor = TabularPredictor(label=label, path=save_path, verbosity=0).fit(train_data)

# basic model evaluation
predictor = TabularPredictor.load(save_path)  # unnecessary, just demonstrates how to load previously-trained predictor from file
y_pred = predictor.predict(test_data_nolab)
print("Predictions:  \n", y_pred)
perf = predictor.evaluate_predictions(y_true=y_test, y_pred=y_pred, auxiliary_metrics=True)

predictor.leaderboard(test_data, silent=True)
```
